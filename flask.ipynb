Q1. What is a RESTful API?

Ans: A RESTful API is a way for two software systems to communicate with each other using the HTTP protocol.
REST stands for Representational State Transfer.
It follows simple rules such as:

it uses URLs to access resources (example: /users, /products/5)

it uses HTTP methods like GET, POST, PUT, DELETE

It Sends and receives data mostly in JSON format

Q4. What is the purpose of the @app.route() decorator in Flask?

Ans: The @app.route() decorator is used to map a URL to a specific function in a Flask application.
When a user visits a particular route (URL), Flask executes the function linked to that route.
Connects URL to Logic:
Each route is linked to a view function that returns a response to the user.
Supports HTTP Methods:
You can specify which methods are allowed for a route.

Q7. How do you return JSON responses in Flask?

Ans: Flask provides the jsonify() function for returning JSON data to the client.
It Automatically converts Python dictionaries to JSON.
It Sets correct MIME type (application/json).
It Handles Unicode and special characters safe.

```python
from flask import jsonify

@app.route('/details')
def details():
    data = {
        "name": "Goutam",
        "course": "API Development",
        "status": "Active"
    }
    return jsonify(data)
```

Q9. How do you handle errors in Flask (e.g., 404)?

Ans: Flask allows custom error handling using the @app.errorhandler() decorator

```python
@app.errorhandler(404)
def page_not_found(error):
    return jsonify({"error": "Page not found"}), 404
```

Q10. How do you structure a Flask app using Blueprints?

Ans: Blueprints help in organizing a Flask application into reusable and modular components.
They are especially useful for large applications.It helps in Organizes project into modules,Makes code clean and maintainable,Allows dividing features into separate files,Supports team collaboration,Avoids one large app.py file

```python
from flask import Blueprint

users = Blueprint('users', __name__)

@users.route('/profile')
def profile():
    return "User Profile Page"
```

```python
from flask import Flask
from users.routes import users

app = Flask(__name__)
app.register_blueprint(users, url_prefix='/user')

if __name__ == '__main__':
    app.run(debug=True)
```
